```python
#CREACIÓN FUNCIONES

#Función IMC
def imc(estatura, peso):
    kilos = peso/2.2
    indice = round(kilos / (estatura ** 2), 2)
    return indice

#Función Grasa corporal
def grasa_corporal(imc, edad, genero):
    gc= (1.20*imc)+(0.23*edad)-(10.8*genero)-5.4
    return round(gc,2)

#Funcion Grasa Magra
def grasa_corporal_magra(peso,  imc):
    kilos = peso/2.2
    
    mcmkg= kilos - (imc*kilos)/100
    return round(mcmkg,2)
```

```python
df = df.assign(IMC=df.apply(lambda row: imc(row['Altura'], row['Peso']), axis=1))

df
```

```python
df = df.assign(grasa_corporal=df.apply(lambda row: grasa_corporal(row['IMC'], row['Edad'], row['genero']), axis=1))

df
```

```python
df = df.assign(grasa_corporal_magra=df.apply(lambda row: grasa_corporal_magra(row['Peso'], row['IMC']), axis=1))

df
```
